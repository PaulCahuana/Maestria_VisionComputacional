```python
#coding=utf-8  
import cv2  
import numpy as np    


import os #librería para crear carpetas
import pandas as pd
from os import listdir
import random
import shutil
from sklearn.model_selection import train_test_split
from glob import glob

#os.mkdir("Nivel_2") #creamos la carpeta para las imágenes
carpetaImages="CK"
carpetaDest=carpetaImages+"_dataset"
os.mkdir(carpetaDest)
carpetas=os.listdir(carpetaImages)

for i in carpetas: 
    os.mkdir(carpetaDest+"/"+i)


for i in carpetas: 
    imgs=os.listdir(carpetaImages+"/"+i)
    for j in imgs:
        
        img = cv2.imread(carpetaImages+"/"+i+"/"+j, 0)  
        print(carpetaImages+"/"+i+"/"+j)
        x = cv2.Sobel(img,cv2.CV_16S,1,0)  
        y = cv2.Sobel(img,cv2.CV_16S,0,1)  
        
        absX = cv2.convertScaleAbs(x)   # Transferencia de regreso a uint8  
        absY = cv2.convertScaleAbs(y)  
        
        dst = cv2.addWeighted(absX,0.5,absY,0.5,0)  
        
        #cv2.imshow("absX", absX)  
        #cv2.imshow("absY", absY)  
        
        #cv2.imshow("Result", dst)  
        backtorgb = cv2.cvtColor(dst,cv2.COLOR_GRAY2RGB)

        cv2.imwrite(carpetaDest+"/"+i+"/"+j, backtorgb)
```
